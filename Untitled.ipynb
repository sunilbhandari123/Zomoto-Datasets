Lets deal with missing values

```python
df.isnull().sum()
```

Here rating and avarage cos has the missing values and tha to are very small we can just 
fill it by median values.

```python
def function(df,variable):
    median=df[variable].median()
    df[variable+'median']=df[variable].fillna(median)
```

```python
function(df,'rate (out of 5)')
```

```python
function(df,'avg cost (two people)')
```

```python
df=df.drop(['rate (out of 5)','avg cost (two people)'],axis=1)
```

```python
df.isnull().sum()
```

Now there are no missing values.

restaurant name,restaurant type,online_order,table booking,cuisines type,area,local address these are

Here all the categorical variable are nominal data so we can use onehot encoding and one hot encoding

```python
col_for_onehotencoding=['online_order','table booking']
```

```python
def fun (df,variable,top_list):  #here we are making a function that will help to convert our top_10 
    for label in top_list:       #categorical items of each variable to numerical variable. 
        df[variable+'_'+str(label)]=np.where(df[variable]==label,1,0)
```

```python
fun(df,'restaurant name',list_10)
```
